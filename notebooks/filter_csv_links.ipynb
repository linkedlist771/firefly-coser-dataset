```python
import pandas as pd
from pathlib import Path
```

```python
ROOT = Path(Path.cwd().parent)
MD_FILES = ROOT / "mds"
CSV_FILES = ROOT / "csvs"
```

```python
CSV_FILE = CSV_FILES / "link_source1.csv"
```

```python
df = pd.read_csv(CSV_FILE)
```

```python
df
```

```python
df.columns
```

```python
## 1. First, we only process the non-video content

# Filter out rows where:
# 1. 链接 contains "xxx" (placeholder links)
# 2. 内容类型 contains "视频" (video content)

df_filtered = df[
    (~df["链接"].str.contains("xxx", na=False))
    & (~df["链接"].str.contains("video", na=False))
    & (~df["内容类型"].str.contains("视频", na=False))
]

print(f"Original rows: {len(df)}")
print(f"Filtered rows: {len(df_filtered)}")
print(f"Removed rows: {len(df) - len(df_filtered)}")
df_filtered.drop(columns=["#"], inplace=True)
df_filtered.reset_index(drop=True, inplace=True)
df_filtered
```

```python
df_filtered.to_csv(CSV_FILES / "link_source1_non_video_filtered.csv", index=False)
```
